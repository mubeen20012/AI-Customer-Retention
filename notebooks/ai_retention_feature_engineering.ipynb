#Feature Engineering

```python
def tenure_group(tenure):
    if tenure <= 12:
        return 'New'
    elif tenure <= 36:
        return 'Mid'
    else:
        return 'Loyal'

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
df['is_long_term_contract'] = df['Contract'].apply(
    lambda x: 0 if x == 'Month-to-month' else 1
)
```

```python
df['is_auto_payment'] = df['PaymentMethod'].apply(
    lambda x: 1 if 'automatic' in x.lower() else 0
)
```

```python
service_cols = [
    'PhoneService', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies'
]

df['num_services'] = df[service_cols].apply(
    lambda row: sum(row == 'Yes'), axis=1
)
```

```python
df['avg_monthly_spend'] = df.apply(
    lambda x: x['TotalCharges'] / x['tenure'] if x['tenure'] > 0 else 0,
    axis=1
)
```

```python
df['is_single'] = df.apply(
    lambda x: 1 if (x['Partner'] == 'No' and x['Dependents'] == 'No') else 0,
    axis=1
)
```

```python
print(df.head())
```

```python
#Category Encoding
cat_encod=['gender' , 'Partner' ,'Dependents' , 'PhoneService',   'PaperlessBilling' ]
for col in cat_encod:
  if col == 'gender':
    df[col]=df[col].map({'Male':1,'Female':0})
  else:
   df[col]= df[col].map({'Yes':1,'No':0})
```

```python
df['Churn']=df['Churn'].map({'Yes':1,'No':0})
```

```python
#one hot Encoding
one_hot=[ 'MultipleLines' ,'InternetService', 'OnlineSecurity', 'DeviceProtection' ,'TechSupport', 'StreamingTV', 'StreamingMovies' , 'Contract','PaymentMethod','tenure_group']
df=pd.get_dummies(df,columns=one_hot,drop_first=True,dtype=int)
```

```python
df=pd.get_dummies(df,columns=['OnlineBackup'],drop_first=True,dtype=int)
```
